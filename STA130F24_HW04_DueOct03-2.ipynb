## The confidence interval is calculated by plus or minus the product of the Standard Error of the Mean Critical Value and the SEM. The Critical value is divided into Z-Distribution and t-Distribution. When the sample size is above 30,we can use Z-Distribution, and its Critical Value for the 95% confidence interval is about 1.96. When the sample size is less than 30, we need use the t-Distribution, and its Critical Value is related to the Degrees of Freedom. Degrees of Freedom is the number of samples minus one, and then we can get the Critical Value according to different Degrees of Freedom.

## The 95% confidence interval means that 95% of the samples we collect are included in the true average, which gives us more confidence in the data we use. When the sample mean plus and minus about 2 times the standard error, it means that the different data in the sample are distributed very widely. Or we need more samples to estimate the average more accurately.

## Quantitative analysis can give accurate results of each value of the sample, which is convenient for us to conduct subsequent analysis

````python
import numpy as np

Mean = 2.9
Std = 2.86
n = 10
z = 1.96

SME = Std / np.sqrt(n)
CIL = (Mean - z * SME)
CIH = (Mean + z * SME)




print("The 95% Confidence intervals of initial health score is: [",CIL,",", CIH,"]")

````
